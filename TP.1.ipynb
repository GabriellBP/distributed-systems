Modifique a implementação do B-multicast para utilizar communicação síncrona.

```julia
# B-multicast Asynchronous.
using Distributed

function B_multicast(g,m,B_deliver)
    for p in g
        remotecall(B_deliver, p, m)
    end
end
```

```julia
# Synchronous B-multicast implementation.
using Distributed

function Sync_B_multicast(g, m, B_deliver)
    for p in g
        remotecall_wait(B_deliver, p, m)
    end
end
```

```julia
# Asynchronous:
B_multicast(g,"oi!",println)
```

```julia
# Synchronous:
Sync_B_multicast(g, "oi!", println)
```

Execute as duas versões do B-multicast (síncrona e assíncrona) e compare as saídas.

A ordem das mensagens das saídas foram as mesmas para ambas versões do B-multicast? Por quê?

**Resposta**: _Não. A versão assíncrona não garante a ordem em que os processos serão executados pois depende do tempo que cada processo levará para terminar para ser enviado ao receptor. Ao contrário da versão síncrona que garante a ordem em que os processos serão executados, ou seja, o receptor irá sempre receber as mensagens na ordem que forem enviadas pelo emissor pois este respeita a ordem de envio_

### Exercício 2: Análise de desempenho do B-multicast (2,0 pontos)

Meça o tempo de execução duas versões do B-multicast (síncrona e assíncrona). Dica: utilize a macro `@time`.

* Execute 50 vezes essa medição e construa um gráfico em forma de barras:
	* cada barra representando cada versão do B-multicast;
	* calcule e coloque no gráfico o [intervalo de confiança](https://en.wikipedia.org/wiki/Confidence_interval) com nível de 95%.
    
**Gráfico**: 

```julia
# executing 50 times the sync and the async B-multicast function
g = workers()  # all workers

# array of the time values
b_time_sync = Float64[] 
b_time_async = Float64[]
for i = 1:50
    push!(b_time_sync, @elapsed Sync_B_multicast(g,"oi!",println))
end
for i = 1:50
    push!(b_time_async, @elapsed B_multicast(g,"oi!",println))
end
```

```julia
# confint(OneSampleTTest(x)) returns the 95% confidence interval for the true population mean, reference:
# -> https://stackoverflow.com/questions/20425473/calculate-confidence-interval-in-julia
# -> https://stackoverflow.com/questions/34241766/julia-language-is-there-a-t-test
# -> https://github.com/JuliaStats/HypothesisTests.jl
# -> https://docs.juliaplots.org/latest/tutorial/
# -> https://github.com/JuliaPlots/StatsPlots.jl
# -> https://gist.github.com/gizmaa/7214002#barplot

using HypothesisTests, Statistics, Plots

(low_sync, high_sync) = confint(OneSampleTTest(b_time_sync))
sync_error = (high_sync - low_sync) / 2
(low_async, high_async) = confint(OneSampleTTest(b_time_async))
async_error = (high_async - low_async) / 2

bar(["Synchronous", "Asynchronous"], [mean(b_time_sync), mean(b_time_async)], labels=["B-Multicast"], ylabel="seconds", yerr=[sync_error, async_error], color="#0f87bf")
```

* Qual das duas versões teve o menor tempo de execução? Por quê?

**Resposta**: _A versão assíncrona, já que os processos são executados em paralelo._

### Exercício 3: Análise de desempenho do R-multicast (3,0 pontos)

Implemente duas versões do algoritmo R-multicast: uma assíncrona e outra síncrona.

* R-multicast síncrono

```julia
# Synchronous R-multicast implementation.

using Distributed

function Sync_R_multicast(g, m, R_deliver)
    for p in g
        if m in msg_received
            remotecall_wait(R_deliver, p, m)
        else  
            push!(msg_received, m)
            Sync_R_multicast(g, m, R_deliver)
        end
    end
end
```

* R-multicast assíncrono

```julia
# Asynchronous R-multicast implementation.

using Distributed

function Async_R_multicast(g, m, R_deliver)    
    for p in g
        if m in msg_received 
            remotecall(R_deliver, p, m)
        else
            push!(msg_received, m)  
            Async_R_multicast(g, m, R_deliver)
        end
    end
end
```

Faça a mesma análise de desempenho para o R-multicast que você fez no Exercício 2.


**Gráfico**: 

```julia
# executing 50 times the sync and the async R-multicast function
@everywhere global msg_received = Set()
g = workers()  # all workers

# array of the time values
r_time_sync = Float64[] 
r_time_async = Float64[]
for i = 1:50
    push!(r_time_sync, @elapsed Sync_R_multicast(g,i,println))
end
for i = 51:100
    push!(r_time_async, @elapsed Async_R_multicast(g,i,println))
end

# Here, I used differents values to the print function, to avoid errors
```

```julia
using HypothesisTests, Statistics, Plots

(low_sync, high_sync) = confint(OneSampleTTest(r_time_sync))
sync_error = (high_sync - low_sync) / 2
(low_async, high_async) = confint(OneSampleTTest(r_time_async))
async_error = (high_async - low_async) / 2

bar(["Synchronous", "Asynchronous"], [mean(r_time_sync), mean(r_time_async)], ylabel="seconds", labels=["R-Multicast"], yerr=[sync_error, async_error], color="#de1738")
```

* Qual das duas versões (R-multicast assíncrono e assíncrono) teve o menor tempo de execução? Por quê?

**Resposta**: A versão assíncrona teve o menor tempo de execução. Isso se dá porque a execução acontece de forma simultânea, em paralelo, não precisa estar alinhada globalmente, com isso realiza as operações apenas verificando a flag, ao contrário do síncrono que precisa estar alinhado globalmente e o receptor de ve checar se recebeu a mensagem antes de poder enviála, ou seja, mesmo devido a essas checagens a mais, no R-Multicast, o assíncrono permanece mais eficiente.

### Exercício 4: Análise de desempenho: B-multicast vs. R-multicast (3,0 pontos)
